## 2. 哪有人一開始就畫圖的

Python 畫圖基本上就是用

    plt.plot(X, Y)
    
其中,

* X: 點的ｘ座標的串列
* Y: 點的 y 座標的串列

X 座標可以省略!

```python
plt.plot([1, 35, 17, 87, 14])
```

好高級的亂數指令

    np.random.randn(n)
    
意思是從標準常態分布中取 n 個點。

```python
plt.plot(np.random.randn(100))
```

畫出 (1.1, 2), (2, -5), (2.8, 10), (3.5, -6), (4, 8)

```python
X = [1.1, 2, 2.8, 3.5, 4]
Y = [2, -5, 10, -6, 8]
```

```python
plt.plot(X, Y)
```

也可以只畫點。

```python
plt.scatter(X, Y)
```

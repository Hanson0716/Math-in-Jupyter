## 冰雹猜想

我們由一個正整數 $a_0$ 開始開始, 每個 $a_n$ 下一個下一個 $a_{n+1}$ 決定的方式是這樣:

$$a_{n+1} =  \begin{cases}
a_n/2, & \mbox{如果 $a_n$ 是偶數}\\
3a_n+1, & \mbox{如果 $a_n$ 是奇數}
\end{cases}$$

冰雹猜想是說, 不管 $a_0$ 是哪個正整數, 最後都能得到 1。

[註] 冰雹猜想正式的名稱是 Collatz conjecture

```python
def hail(k):
    p = [k]
    while k != 1:
        if k%2 == 0:
            k = k/2
        else:
            k = 3*k + 1
        p.append(k)
        
    plt.plot(p)
    print(p)
```

```python
hail(3)
```

```python
hail(27)
```

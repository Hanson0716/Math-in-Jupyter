## 參數式畫圖

提醒極座標:

$$\begin{cases}
x = r \cos \theta \\
y = r \sin \theta
\end{cases}$$

記得可愛的 Python 3 可以用 $\theta$ 當變數, 而 Jupyter Notebook 可以用可以用

    \theta + TAB 鍵
    
打出 $\theta$ 來!!

### 畫一個圓

```python
θ = np.linspace(-2*np.pi, 2*np.pi, 200)
```

```python
r = 3
x = r*np.cos(θ)
y = r*np.sin(θ)
```

```python
plt.plot(x,y)
```

### 如果堅持要修正比率...

```python
ax = plt.gca()
ax.set_aspect('equal')

plt.plot(x,y)
```

### 任何的參數式都很方便

注意, 只改了 R 的定義。

```python
θ = np.linspace(-2*np.pi, 2*np.pi, 200)

r = 1-np.sin(θ)
x = r*np.cos(θ)
y = r*np.sin(θ)
```

```python
ax = plt.gca()
ax.set_aspect('equal')

plt.plot(x,y)
```

### 讓同學想, 怎麼畫出這類型的圖?

```python
θ = np.linspace(-2*np.pi, 2*np.pi, 1000)

r = 2 + 0.1*np.sin(25*θ)
x = r*np.cos(θ)
y = r*np.sin(θ)
```

```python
ax = plt.gca()
ax.set_aspect('equal')

plt.plot(x,y)
```

## 互動的複習

記得有函數就有互動!

```python
from ipywidgets import interact
```

順便介紹一下, Python 可以設函數的預設值!

```python
θ = np.linspace(-2*np.pi, 2*np.pi, 1000)

def myplot(k=1):
    r = 2 + 0.1*np.sin(k*θ)
    x = r*np.cos(θ)
    y = r*np.sin(θ)


    ax = plt.gca()
    ax.set_aspect('equal')

    plt.plot(x,y)
```

```python
interact(myplot, k=(1,50))
```

## 介紹怎麼畫分段定義函數

之前有同學問, 我原本回答的方式不太好...

假設要畫這個分段定義函數

$$f(x) = \begin{cases}
x^2 & x<0,\\
0 & x \geq 0
\end{cases}$$

```python
x = np.linspace(-5,5,200)
```

```python
y = np.piecewise(x, [x<0, x>=0], [lambda x: x**2, 0])
```

```python
plt.plot(x,y)
```

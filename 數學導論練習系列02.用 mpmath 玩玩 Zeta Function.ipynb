## 使用 `mpmath` 套件

這裡我們使用 `mpmath`, 最大的原因就是它很會算 Riemann Zeta Function, 實數的也會, 複數也會!

```python
ζ = mp.zeta
```

```python
ζ(2+3j)
```

說明一下:
    
* `mpc`: 表 `mpmath` 的複數
* `mpf`: 表 `mpmath` 的浮點數, 也就是一般實數在電腦的表示

這樣有點醜, 是不是可以就像一般 Python 顯示數字就好, 不要這樣嚇人?

## 看見 Riemann Zeta Function

我們來看看 critical line $\mathcal{R}(z) = 0.5$ 會被映到哪裡呢?

```python
x = []
y = []
cl = np.linspace(-50,50,2000)

for t in cl:
    p = ζ(0.5 + t*1j)
    x.append(p.real)
    y.append(p.imag)
    
plt.plot(x,y,alpha=0.5)

# 畫出原來的 critical line
plt.plot([0.5, 0.5], [min(y)-0.5, max(y)+0.5])

# 畫出原點
plt.scatter([0], [0], s=100, c='r')
```

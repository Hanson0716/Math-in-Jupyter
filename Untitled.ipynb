```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import sympy as sym
```

```python
θ = np.linspace(0, 2*np.pi, 500)
```

```python
x = np.cos(θ)
y = np.sin(θ)
```

```python
k = 87

ax = plt.gca()
ax.set_aspect('equal')

plt.plot(x, y, zorder=1)
plt.scatter(x[k], y[k], c='r', s=100, zorder=2)
```

```python
def myplot(k):
    ax = plt.gca()
    ax.set_aspect('equal')

    plt.plot(x,y)
    plt.scatter(x[k], y[k], c='r', s=100)
```

```python
from ipywidgets import interact
```

```python
interact(myplot, k=(0,499))
```

```python
import pyaudio
```

```python
fs = 44100
duration = 1.0
f = 440.0
```

```python
t = np.linspace(0, duration, fs)
```

```python
y = np.sin(2*np.pi*f*t).astype(np.float32)
```

```python
plt.plot(t,y)
plt.xlim(0, 0.02)
```

```python
samples = y.tobytes()
```

```python
p = pyaudio.PyAudio()
```

```python
stream = p.open(format=pyaudio.paFloat32,
               channels=1,
               rate=fs,
               output=True)
```

```python
stream.write(samples)
```

```python
stream.stop_stream()
stream.close()
```

```python
p.terminate()
```

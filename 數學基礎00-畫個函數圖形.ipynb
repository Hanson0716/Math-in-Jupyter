### 1. 讀入標準套件

所有程式語言都需要套件「加持」, 才能夠發揮我們想要的功能。對我們來說, 有三大套件基本上是必要的, 也就是每次你一開始寫程式都讀進來:

* `numpy`: 數值型的計算
* `sympy`: 符號型的計算
* `matplotlib`: Python 的標準畫圖套件



```python
%matplotlib inline

import numpy as np
import sympy as sp
import matplotlib.pyplot as plt
```

### 2. 畫個函數

畫函數基本上就是描點! Python (事實上是 `matplotlib`) 喜歡把 x 和 y 座標分開。假設我們要畫的函數是:

$$f(x) = \mathrm{sinc}(x)$$

首先我們先決定 x 的範圍, 比如說 $[-5, 5]$, 然後我們要決定需要描多少點, 比如說 300 個點。我們的指令會像這樣

    x = np.linspace(起始值, 終點, 點數)
    
然後 x 就有我們要的點數。之後, `numpy` 強強的地方, 就是可以一次算完!

    y = np.sinc(x)

```python
x = np.linspace(-5, 5, 300)
y = np.sinc(x)
```

接著下來, 我們只要用

    plt.plot(x, y)

就可以把點一個個點出來, 再連起來!

```python
plt.plot(x, y)
```

### 3. 參數式的畫圖

現在我們來畫參數式的圖。我們會發現基本上都是一樣的, 也就是我們有 x, y 參數式:

    x(t)
    y(t)
    
接著, 我們只要用 `np.linspace` 把 t 在指定範圍切出若干點, 然後代入算出 x, y 的座標。再用 `np.plot` 畫出來。

一個常用的參數化方法是用「極座標」, 比如說, 一個半徑是 3 的圓極座標式子是簡簡單單的:

$$r = 3$$

但是, Python 不太會極座標, 我們要換成直角座標系統:

$$x = r\cos(\theta)$$
$$y = r\sin(\theta)$$

我們順變來給大家看看一個耍寶技巧, 這是 $\LaTeX$ 數學符號打法, 可直接用在 Jupyter Notebook。比如說, 我們想打出 $\pi$, 只要:

    \pi
    
再按 `Tab` 就可以了!

```python
π = np.pi
```

```python
π
```

這是不是很炫? $\pi$ 真的是 $\pi$!!

算然 $\theta$ 也可以打出來, 然後我們想在 $[0, 2\pi]$ 區間, 切出 300 個點。

```python
θ = np.linspace(0, 2*π, 300)
```

接著我們要畫的是 $r=3$ 這個極座標式的圖。

```python
r = 3
```

代入 x, y 標準計算方式。

```python
x = r*np.cos(θ)
y = r*np.sin(θ)
```

再畫出來!

```python
plt.plot(x,y)
```

耶! 成功畫出來了! 可是, 怎麼是橢圓呢? 原來是 x, y 每單位長度是不一樣的! 我們可以調整一下...

### 4. 極座標的進階想法

我們來重畫圓, 但是要畫出正圓。

```python
θ = np.linspace(0, 2*π, 300)
```

```python
r = 3
```

```python
x = r*np.cos(θ)
y = r*np.sin(θ)
```

要畫出來有點複雜的是, 我們需要知道現在在畫圖的這個 axes (也就是這張圖)。要做到這個要用

    plt.gca()
    
再來就設 x, y 一致的單位長度。

```python
ax = plt.gca()

plt.plot(x, y)
ax.set_aspect('equal')
```

果然有圓了!

我們再來做更有趣的實驗, 就是 $r$ 也不是固定不動, 而是依不同的 $\theta$ 有不同的長度。會畫出什麼來呢? 我們看一個例子。

```python
k=10
r = 3 + 0.5*np.sin(k*θ)
```

```python
x = r*np.cos(θ)
y = r*np.sin(θ)
```

```python
ax = plt.gca()
ax.set_aspect('equal')

plt.plot(x, y)
```

最後其實我們只是為了耍寶, 介紹經過神秘指令
